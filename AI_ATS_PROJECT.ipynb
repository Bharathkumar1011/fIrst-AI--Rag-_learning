```python
#Intialzng faker to genrerate randon names and data
fake = faker.Faker()
```

```python

# Predefined sets of skills, degrees, job titles, and companies
skills = ['Python', 'Machine Learning', 'Data Analysis', 'SQL', 'Communication', 'Team Leadership', 'Java', 'C++', 'Project Management']
job_titles = ['Data Scientist', 'Software Engineer', 'Project Manager', 'Business Analyst', 'DevOps Engineer']
degrees = ['B.Sc. Computer Science', 'B.Sc. Information Technology', 'M.Sc. Data Science', 'MBA', 'B.A. Business Administration']
companies = ['TechCorp', 'DataGen', 'Innovate Inc.', 'NextGen Solutions', 'Global Tech']
```

```python
# Function to generate a synthetic resume
def generate_resume():
    resume = {
        "name": fake.name(),
        "contact_info": fake.email(),
        "education": {
            "degree": random.choice(degrees),
            "university": fake.company(),
            "year_of_completion": random.randint(2010, 2023)
        },
        "experience": [
            {
                "job_title": random.choice(job_titles),
                "company": random.choice(companies),
                "years_of_experience": random.randint(1, 10),
                "description": "Worked on various projects including AI, data analysis, and software development."
            }
        ],
        "skills": random.sample(skills, random.randint(3, 6)),
        "certifications": random.sample(['AWS Certified', 'PMP', 'Google Data Engineer', 'None'], 1)
    }
    return resume
```

```python
# Generate multiple resumes
synthetic_resumes = [generate_resume() for _ in range(50)]
```

```python
# Predefined job descriptions
job_titles = ['Data Scientist', 'Software Engineer', 'Project Manager', 'Business Analyst', 'DevOps Engineer']
required_skills = {
    'Data Scientist': ['Python', 'Machine Learning', 'Data Analysis', 'SQL'],
    'Software Engineer': ['Java', 'C++', 'SQL', 'Team Leadership'],
    'Project Manager': ['Project Management', 'Communication', 'Team Leadership'],
    'Business Analyst': ['Data Analysis', 'Communication', 'SQL'],
    'DevOps Engineer': ['Python', 'Cloud Infrastructure', 'SQL', 'DevOps Tools']
}
```

```python
# Function to generate a synthetic job description
def generate_job_description():
    job_title = random.choice(job_titles)
    job_description = {
        "job_title": job_title,
        "company": fake.company(),
        "location": fake.city(),
        "required_skills": required_skills[job_title],
        "experience_required": random.randint(1, 10),
        "preferred_education": random.choice(degrees)
    }
    return job_description
```

```python
# Generate multiple job descriptions
synthetic_job_descriptions = [generate_job_description() for _ in range(10)]
```

```python
# Print or save the data
jobs = []
for job in synthetic_job_descriptions:
    jobs.append(job)
    print(job)
```
